```python
# LIST OF GRAPH GENERATORS --> add
# Need to find one that makes a connected graph somehow

import networkx as nx
seed=1000                                           # seed the graph for reproducibility
G= nx.gnp_random_graph (200, .01, seed=seed )       # here we create a random binomial graph with 20 nodes and an average (expected) connectivity of 10*.3= 3. <-- you can look at the example, these numbers didn't make sense from the start, but 200 * .01 = connectivity of 2, im assuming


makeConnected = list(nx.k_edge_augmentation(G, k=4))
G.add_edges_from(makeConnected)
print("Connectivity = ", end = '')
print(nx.edge_connectivity(G))
print("Is connected = ", end = '')
print(nx.is_connected(G))
print ( G.nodes() )
```
